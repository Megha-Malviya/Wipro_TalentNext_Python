```python
# Text Preprocessing on SMSSpamCollection Dataset

# Step 1: Import libraries
import pandas as pd
import numpy as np
import re
import string

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download necessary nltk data
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# Step 2: Load dataset
# Ensure you have SMSSpamCollection dataset in your working directory
# Dataset format: label <tab> message
df = pd.read_csv("SMSSpamCollection", sep='\t', names=["label", "message"])

# Step 3: Check dataset
print(df.head())
print(df.info())
print(df['label'].value_counts())

# Step 4: Encode labels
# ham = 0, spam = 1
df['label_num'] = df['label'].map({'ham':0, 'spam':1})

# Step 5: Text Preprocessing function
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Tokenize
    tokens = nltk.word_tokenize(text)
    # Remove stopwords and lemmatize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)

# Step 6: Apply preprocessing
df['clean_message'] = df['message'].apply(preprocess_text)

# Step 7: Check processed data
print(df[['message','clean_message']].head(10))

# Step 8: Save the clean dataset
df.to_csv("SMSSpamCollection_Clean.csv", index=False)

print("✅ Preprocessing completed and saved to SMSSpamCollection_Clean.csv")
```

```python
# -----------------------------
# 1. Import Libraries
# -----------------------------
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import word_tokenize, ne_chunk, pos_tag

# Download required nltk data (only first time)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')

# -----------------------------
# 2. Load Dataset
# -----------------------------
df = pd.read_csv("SMSSpamCollection", sep="\t", names=["label", "message"])

print("✅ Dataset Loaded")
print(df.head())

# -----------------------------
# 3. Define Preprocessing Function
# -----------------------------
def preprocess_text(text):
    # Step 1: Lowercasing
    text = text.lower()
    
    # Step 2: Remove special characters & digits
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Step 3: Tokenization
    tokens = word_tokenize(text)
    
    # Step 4: Stopword Removal
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if w not in stop_words]
    
    # Step 5: Stemming
    stemmer = PorterStemmer()
    stemmed = [stemmer.stem(w) for w in tokens]
    
    # Step 6: Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(w) for w in stemmed]
    
    return " ".join(lemmatized)

# -----------------------------
# 4. Apply Preprocessing
# -----------------------------
df['clean_message'] = df['message'].apply(preprocess_text)

print("\n✅ Sample after preprocessing:\n")
print(df[['message', 'clean_message']].head())

# -----------------------------
# 5. Named Entity Recognition (NER)
# -----------------------------
def extract_entities(text):
    tokens = word_tokenize(text)
    tagged = pos_tag(tokens)
    entities = ne_chunk(tagged, binary=True)  # binary=True => PERSON/ORG/LOCATION
    return entities

print("\n✅ Example NER Output:\n")
sample_text = df['message'][0]
print("Original:", sample_text)
print("Entities:", extract_entities(sample_text))
```
