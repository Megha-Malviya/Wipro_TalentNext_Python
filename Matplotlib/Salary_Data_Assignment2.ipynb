```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("Salary.csv")
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Univariate Analysis
sns.histplot(df["YearsExperience"], bins=10, kde=True, color="orange")
plt.title("Years of Experience Distribution")
plt.show()
```

```python
sns.histplot(df["Salary"], bins=10, kde=True, color="green")
plt.title("Salary Distribution")
plt.show()
```

```python
# Bivariate Analysis
sns.scatterplot(x="YearsExperience", y="Salary", data=df, color="red")
plt.title("Years of Experience vs Salary")
plt.show()
```

```python




sns.regplot(x="YearsExperience", y="Salary", data=df, color="blue")
plt.title("Regression Plot: Experience vs Salary")
plt.show()






```
