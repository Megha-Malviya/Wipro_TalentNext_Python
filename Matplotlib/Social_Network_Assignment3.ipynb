```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("Social_Network_Ads.csv")
print(df.head())
print(df.info())
print(df.describe())
```

```python

# Univariate Analysis
sns.countplot(x="Gender", data=df, palette="pastel")
plt.title("Gender Distribution")
plt.show()

```

```python
sns.histplot(df["Age"], bins=20, kde=True, color="purple")
plt.title("Age Distribution")
plt.show()


```

```python
# Bivariate Analysis
sns.scatterplot(x="Age", y="EstimatedSalary", hue="Purchased", data=df, palette="coolwarm")
plt.title("Age vs Estimated Salary with Purchase Status")
plt.show()

```

```python
sns.boxplot(x="Purchased", y="Age", data=df, palette="Set3")
plt.title("Age vs Purchase Decision")
plt.show()
```
